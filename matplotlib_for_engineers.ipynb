```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Line PLOTS
### Making Plots Actually Look Nice

```python
x = np.linspace(0,15,10)
y = np.sin(x) + 0.1*np.random.randn(len(x))
x,y
```

```python
plt.plot(x,y)
plt.grid() # to show the grid lines...
plt.show()
```

```python
## different forms of plotting the lines
```

```python
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'o--') # the data points are displayed on the chart
```

```python
plt.plot(x,y,'o--', color = 'aqua')
```

```python
plt.plot(x,y,'o--', lw = 5, ms = 10) # lw for line width and ms for point size
```

### figsize

```python
plt.figure(figsize=(5,4))
plt.plot(x,y,'o--')
```

## LABLES

```python
plt.figure(figsize = (8,5))
plt.grid()
plt.plot(x,y,'o--', color='cyan',lw = 1, ms = 4, label = 'Infinite curve')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(fontsize=10)
```

x and y limits, multiple line plots together

```python
x2 = np.linspace(0,15,10)
y2 = np.sin(x2)
x2,y2
```

```python
plt.plot(x2,y2)
```

```python
plt.figure(figsize = (10,7))
plt.grid() # show the grid
plt.plot(x,y,'o--', color='purple',label='Data 1')
plt.plot(x2,y2, '-', color='aqua',label='Theory')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.ylim(top=1.5) # extending the top part to accomade the legends
plt.legend(fontsize=20)
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.grid() # show the grid
plt.plot(x,y,'o--', color='purple',label='Data 1')
plt.plot(x2,y2, '-', color='aqua',label='Theory')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.ylim(top=1.5) # extending the top part to accomade the legends
plt.legend(fontsize=20, ncol=2) # 2 columns for legend
plt.show()
```

### Histogram

```python
hist_data = np.random.randn(1000)*0.2 + 0.4
hist_data
```

```python
## Quick Histogram
```

```python
plt.hist(hist_data, bins=30)
plt.show()
```

```python
plt.hist(hist_data, bins=100) # increasing the bins makes the bars narrow
plt.show()
```

### Plotting a density plot where the area is 1 from a histogram

```python
plt.figure(figsize=(10,8))
plt.hist(hist_data, bins=30,density=True)
plt.grid()
plt.show()
```

```python
hist_data2 = np.random.randn(1000)*0.2 + 0.4
```

```python
plt.figure(figsize=(10,8))
plt.hist(hist_data, bins=30,density=True, histtype='step', color='green')
plt.hist(hist_data2, bins=30,density=True, histtype='step', color='red')
plt.grid()
plt.show()
```

### using a more advanced API in matplotlib

#### Single Axes in one figure

```python
fig, ax = plt.subplots(1,figsize=(12,4))
ax.plot(x,y,'o--')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.grid() # to show the grid
```

### Discontinuation

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x, np.sin(x),'--', label='sin of x')
plt.title('Sine Wave Graph')
plt.xlabel('X values')
plt.ylabel('Y values')
# adding the upper limit for y
plt.ylim(top=1.5)
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
