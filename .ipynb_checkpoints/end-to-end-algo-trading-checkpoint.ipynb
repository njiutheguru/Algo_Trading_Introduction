```python
 # Focusing on the returns of apple only..
```

```python
apple = close.AAPL.copy().to_frame()
```

```python
apple
```

```python
# calculate the decrease/ increase from the previous day...
```

```python
apple.shift(periods=1)
```

```python
# create a new column
apple['lag1'] = apple.shift(periods=1)
apple
```

```python
apple.AAPL.sub(apple.lag1)
```

```python
apple['Diff'] = apple.AAPL.sub(apple.lag1)
apple
```

```python
# calculating the percentage change...
```

```python
apple['% change'] = apple.AAPL.div(apple.lag1).sub(1).mul(100)
apple
```

```python
# using the diff method as well
apple['Diff2'] = apple.AAPL.diff(periods=1)
apple
```

```python
apple['%change2'] = apple.AAPL.pct_change(periods=1).mul(100)
apple
```

```python
# delete the columns...
```

```python
del apple['Diff2']
apple
```

```python
del apple['% change']
apple
```

```python
# changing the name of the functions...
apple.columns
```

```python
apple.rename(columns={'%change2':'change'}, inplace = True)
apple
```

```python
apple
```

```python
apple.AAPL.resample('M').last()
```

```python
# resample using the last business day 'BM'
```

```python
apple.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
# calculating the mid-return and std...
```

```python
del apple['change']
apple
```

```python
ret = apple.pct_change().dropna()
```

```python
ret
```

```python
ret.info()
```

```python
ret.plot(kind='hist',figsize=(12,8), bins = 100)
plt.show()
```

```python
# calculating the mid-return...
```

```python
daily_mean_ret = ret.mean()
daily_mean_ret
```

```python
var_daily = ret.var()
var_daily
```

```python
std_daily = np.sqrt(var_daily)
std_daily
```

```python
ret.std() # alternative way
```

```python
 # annual_mean_return

annual_mean_ret = daily_mean_ret * 252
annual_mean_ret
```

```python
annual_var_return = var_daily * 252
annual_var_return
```

```python
annual_std_return = np.sqrt(annual_var_return)
annual_std_return
```

```python
ret.std() * np.sqrt(252) # alternative way
```
